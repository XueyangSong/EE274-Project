```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
algorithms_comparisons_long_fse = pd.read_csv("result/algorithms_comparisons_long.csv", index_col=0)
algorithms_comparisons_long_fse.sort_values(["File Name","Compression Rate"])
```

```python
prefix_length_tuning_crooked = pd.read_csv("result/prefix_length_tuning_crooked.csv", index_col=0)
prefix_length_tuning_crooked = prefix_length_tuning_crooked[prefix_length_tuning_crooked["Parsing"]=="optimal"].sort_values(["File Name","Compression Rate"])
prefix_tuning_suc_1 = prefix_length_tuning_crooked.head(5)[["File Name", "Hash Prefix Length", "Compression Rate"]]
prefix_tuning_suc_1 = prefix_tuning_suc_1.sort_values(["Hash Prefix Length"])
prefix_tuning_suc_1
```

```python
prefix_length_tuning_sof = pd.read_csv("result/prefix_length_tuning_sof.csv", index_col=0)
prefix_length_tuning_sof = prefix_length_tuning_sof[prefix_length_tuning_sof["Parsing"]=="optimal"].sort_values(["File Name","Compression Rate"])
prefix_tuning_suc_2 = prefix_length_tuning_sof.head(5)[["File Name", "Hash Prefix Length", "Compression Rate"]]
prefix_tuning_suc_2 = prefix_tuning_suc_2.sort_values(["Hash Prefix Length"])
prefix_tuning_suc_2
```

```python
plt.plot(prefix_tuning_suc_1["Hash Prefix Length"], prefix_tuning_suc_1["Compression Rate"], 'bo--', label="Crooked Man.txt")
plt.plot(prefix_tuning_suc_2["Hash Prefix Length"], prefix_tuning_suc_2["Compression Rate"], 'ro--', label="Sign of Four.txt")
plt.title("Hash Prefix Length Tuning")
plt.xlabel("Hash Prefix Length")
plt.xticks(np.arange(5, 10, 1))
plt.ylabel("Compression Rate")
plt.legend(loc='best')
plt.show()
plt.savefig("result/prefix_length_tuning.png")
```

```python
prefix_search_range_tuning_sof = pd.read_csv("result/prefix_search_range_tuning_sof.csv", index_col=0)
prefix_search_range_tuning_sof = prefix_search_range_tuning_sof.sort_values(["File Name","Compression Rate"])
prefix_search_range_tuning_sof_suc = prefix_search_range_tuning_sof[prefix_search_range_tuning_sof["Parsing"]=="optimal"].head(10).sort_values(['Prefix Search Range'])
prefix_search_range_tuning_sof_suc
```

```python
plt.plot(prefix_search_range_tuning_sof_suc[prefix_search_range_tuning_sof_suc["Hash Prefix Length"]==6]["Prefix Search Range"], prefix_search_range_tuning_sof_suc[prefix_search_range_tuning_sof_suc["Hash Prefix Length"]==6]["Compression Rate"], 'bo--', label="Prefix Length: 6")
plt.plot(prefix_search_range_tuning_sof_suc[prefix_search_range_tuning_sof_suc["Hash Prefix Length"]==5]["Prefix Search Range"], prefix_search_range_tuning_sof_suc[prefix_search_range_tuning_sof_suc["Hash Prefix Length"]==5]["Compression Rate"], 'ro--', label="Prefix Length: 5")
plt.plot(prefix_search_range_tuning_sof_suc[prefix_search_range_tuning_sof_suc["Hash Prefix Length"]==7]["Prefix Search Range"], prefix_search_range_tuning_sof_suc[prefix_search_range_tuning_sof_suc["Hash Prefix Length"]==7]["Compression Rate"], 'yo--', label="Prefix Length: 7")
plt.title("Hash Prefix Search Range Comparisons")
plt.xlabel("Prefix Search Range")
plt.xticks([8, 16, 32])
plt.ylabel("Compression Rate")
plt.yticks(np.arange(0.503, 0.510, 0.001))
plt.legend(loc='best')
plt.show()
plt.savefig("result/prefix_search_range.png")
```
